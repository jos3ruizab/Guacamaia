```python
# --- Carga básica ---
import pandas as pd
import numpy as np
from pathlib import Path
# cambia la ruta si es necesario:
csv_path = r"C:\Users\Jose Ruiz\Documents\Scool\Master serio\Guacamaia\Guacamaia_data\Google_API\google_reviews_with_places_clean.csv"

# archivo de log .txt
log_file_path = Path(r"C:\Users\Jose Ruiz\Documents\Scool\Master serio\Guacamaia\Guacamaia_data\Google_API\proceso.txt")

def log_step(msg):
    log_line = f"[STEP] {msg}"
    # mostrar en consola
    print(log_line)
    # guardar en archivo .txt (append)
    with open(log_file_path, "a", encoding="utf-8") as f:
        f.write(log_line + "\n")

log_step("Cargar dataset de Google Reviews y mostrar información inicial (forma, columnas, tipos y ejemplos de filas)")

# lectura simple 
df = pd.read_csv(csv_path, low_memory=False)

# --- Vistazo rápido ---
print("Filas, Columnas:", df.shape)
print("\nColumnas:")
print(df.columns.tolist())

print("\nTipos de datos:")
print(df.dtypes)

print("\nPrimeras 3 filas:")
display(df.head(3))

print("\nÚltimas 3 filas:")
display(df.tail(3))

# Ajuste opcional para ver textos más largos sin cortar
pd.set_option("display.max_colwidth", 200)
```

```python
log_step("Revisar valores nulos y cardinalidad (valores únicos) por columna en el dataset de Google Reviews")
# --- Resumen de nulos ---
print("Valores nulos por columna:\n")
print(df.isnull().sum())

print("\n--- Cardinalidad (valores únicos por columna) ---")
print(df.nunique())
```

```python
import matplotlib.pyplot as plt
log_step("Visualizar distribución de ratings (1-5 estrellas) y de idiomas solicitados en el dataset de Google Reviews")


# --- Distribución de ratings ---
df['review_rating'].value_counts().sort_index().plot(
    kind='bar',
    color='skyblue',
    edgecolor='black'
)
plt.title("Distribución de ratings (1-5 estrellas)")
plt.xlabel("Rating")
plt.ylabel("Número de reseñas")
plt.show()

# --- Distribución de idiomas ---
df['review_language_requested'].value_counts().plot(
    kind='bar',
    color='lightgreen',
    edgecolor='black'
)
plt.title("Distribución de idiomas solicitados")
plt.xlabel("Idioma")
plt.ylabel("Número de reseñas")
plt.show()
```

```python
log_step("Convertir review_time a fecha, revisar rango temporal y graficar evolución mensual de reseñas")

# --- Convertir review_time a fecha ---
df['review_time'] = pd.to_datetime(df['review_time'], errors='coerce')

# --- Revisar rango temporal ---
print("Reseñas desde:", df['review_time'].min())
print("Reseñas hasta:", df['review_time'].max())

# --- Agrupamos por mes ---
reviews_per_month = df.resample("M", on="review_time").size()

# --- Gráfica ---
reviews_per_month.plot(
    kind="line",
    figsize=(10,4),
    marker="o",
    color="orange"
)
plt.title("Evolución temporal de reseñas (por mes)")
plt.xlabel("Fecha")
plt.ylabel("Número de reseñas")
plt.grid(True)
plt.show()
```

```python
log_step("Crear tabla pivote año-mes y generar heatmap del número de reseñas")

import seaborn as sns

# --- Extraer año y mes ---
df['year'] = df['review_time'].dt.year
df['month'] = df['review_time'].dt.month

# --- Crear tabla pivote: filas=año, columnas=mes ---
pivot = df.pivot_table(
    index='year',
    columns='month',
    values='review_text',
    aggfunc='count'
).fillna(0)   # reemplaza NaN por 0

# --- Heatmap ---
plt.figure(figsize=(10,6))
sns.heatmap(pivot, annot=True, fmt="g", cmap="YlOrRd", cbar=True)
plt.title("Número de reseñas por Año y Mes")
plt.xlabel("Mes")
plt.ylabel("Año")
plt.show()
```

```python
log_step("Calcular y graficar la evolución del rating promedio por mes")

# --- Rating promedio por mes ---
rating_per_month = df.resample("M", on="review_time")['review_rating'].mean()

plt.figure(figsize=(10,4))
rating_per_month.plot(marker="o", color="purple")
plt.title("Evolución del rating promedio por mes")
plt.xlabel("Fecha")
plt.ylabel("Rating promedio")
plt.grid(True)
plt.show()
```

```python
log_step("Crear tabla pivote año-mes con rating promedio y visualizar heatmap")

# --- Pivot con promedio de ratings ---
pivot_rating = df.pivot_table(
    index='year',
    columns='month',
    values='review_rating',
    aggfunc='mean'
).round(2)

# --- Heatmap ---
plt.figure(figsize=(10,6))
sns.heatmap(pivot_rating, annot=True, fmt=".2f", cmap="RdYlGn", center=3)
plt.title("Rating promedio por Año y Mes")
plt.xlabel("Mes")
plt.ylabel("Año")
plt.show()
```

```python
log_step("Crear subset geográfico con lat/lng, convertir a numérico, filtrar valores válidos y revisar filas geolocalizadas")
# Subset de columnas geográficas
geo = df[['place_id','place_name','lat','lng']].copy()

# Asegurar tipo numérico y filtrar rangos válidos
geo['lat'] = pd.to_numeric(geo['lat'], errors='coerce')
geo['lng'] = pd.to_numeric(geo['lng'], errors='coerce')
geo = geo.dropna(subset=['lat','lng'])
geo = geo[(geo['lat'] >= -90) & (geo['lat'] <= 90) & (geo['lng'] >= -180) & (geo['lng'] <= 180)]

print("Filas geolocalizadas:", len(geo))
geo.head(3)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(geo['lng'], geo['lat'], s=5, alpha=0.3)  # lng en eje X, lat en eje Y
plt.title("Distribución geográfica de reseñas (puntos)")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
hb = plt.hexbin(geo['lng'], geo['lat'], gridsize=40, mincnt=1)
plt.title("Densidad de reseñas (hexbin)")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
cb = plt.colorbar(hb)
cb.set_label('Nº reseñas')
plt.show()
```

```python
log_step("Generar mapa de calor con Folium a partir de coordenadas lat/lng de reseñas")

!pip install folium
import folium
from folium.plugins import HeatMap

# Centro del mapa: media de coordenadas
center_lat = geo['lat'].mean()
center_lng = geo['lng'].mean()

m = folium.Map(location=[center_lat, center_lng], zoom_start=6)  # ajusta zoom_start si hace falta

# Lista de puntos [lat, lng]
heat_data = geo[['lat','lng']].values.tolist()

HeatMap(
    heat_data,
    radius=12,     # tamaño del “bloom”
    blur=15,       # suavizado
    max_zoom=12
).add_to(m)

m
```

```python
log_step("Generar mapa de calor con marcadores para los Top 20 lugares con más reseñas")

# Top 20 lugares por nº de reseñas
top_places = df.groupby(['place_id','place_name','lat','lng']).size().reset_index(name='n_reviews')
top_places = top_places.sort_values('n_reviews', ascending=False).head(20)

m2 = folium.Map(location=[center_lat, center_lng], zoom_start=6)

# Capa de calor
HeatMap(heat_data, radius=12, blur=15, max_zoom=12).add_to(m2)

# Marcadores de top lugares
for _, row in top_places.iterrows():
    folium.CircleMarker(
        location=[row['lat'], row['lng']],
        radius=6,
        popup=f"{row['place_name']} — {row['n_reviews']} reseñas",
        fill=True
    ).add_to(m2)

m2
```

```python
log_step("Extraer ciudad desde 'place_address', calcular Top 15 por nº de reseñas y graficar en barras")
# --- Extraer ciudad de la dirección ---
# asumimos que la ciudad está en la penúltima parte de la dirección
df['city'] = df['place_address'].apply(lambda x: str(x).split(",")[-2].strip() if "," in str(x) else x)

# --- Top 15 ciudades por nº de reseñas ---
top_cities = df['city'].value_counts().head(15)

# --- Bar plot ---
plt.figure(figsize=(10,5))
top_cities.plot(kind='bar', color='teal', edgecolor='black')
plt.title("Top 15 ciudades con más reseñas")
plt.xlabel("Ciudad")
plt.ylabel("Número de reseñas")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
log_step("Contar número de reseñas por idioma solicitado y graficar en barras")

# --- Top idiomas ---
lang_counts = df['review_language_requested'].value_counts()

# --- Bar plot ---
plt.figure(figsize=(8,5))
lang_counts.plot(kind='bar', color='orange', edgecolor='black')
plt.title("Número de reseñas por idioma solicitado")
plt.xlabel("Idioma")
plt.ylabel("Número de reseñas")
plt.xticks(rotation=45)
plt.show()
```

```python
log_step("Calcular y graficar el rating promedio por idioma solicitado")

# Rating promedio por idioma
avg_rating_lang = df.groupby('review_language_requested')['review_rating'].mean().sort_values(ascending=False)

# Bar plot
plt.figure(figsize=(8,5))
avg_rating_lang.plot(kind='bar', color='lightblue', edgecolor='black')
plt.title("Rating promedio por idioma")
plt.xlabel("Idioma")
plt.ylabel("Rating promedio")
plt.ylim(0,5)
plt.show()

avg_rating_lang
```

```python
log_step("Analizar evolución mensual de reseñas por idioma y graficar líneas de tendencia")

# Conteo mensual por idioma
reviews_per_month_lang = df.groupby(
    [pd.Grouper(key="review_time", freq="M"), "review_language_requested"]
).size().reset_index(name="n_reviews")

# Pivot para gráfico
pivot_lang = reviews_per_month_lang.pivot(index="review_time", columns="review_language_requested", values="n_reviews").fillna(0)

# Line plot
pivot_lang.plot(figsize=(12,6))
plt.title("Evolución de reseñas por idioma (mensual)")
plt.xlabel("Fecha")
plt.ylabel("Número de reseñas")
plt.legend(title="Idioma")
plt.grid(True)
plt.show()
```

```python
log_step("Calcular longitud promedio de reseñas (en palabras) por idioma y graficar en barras")

# Longitud de reseña (número de palabras en texto original)
df['review_length'] = df['review_original_text'].astype(str).apply(lambda x: len(x.split()))

# Promedio por idioma
avg_len_lang = df.groupby('review_language_requested')['review_length'].mean().sort_values(ascending=False)

plt.figure(figsize=(8,5))
avg_len_lang.plot(kind='bar', color='coral', edgecolor='black')
plt.title("Longitud promedio de reseñas por idioma")
plt.xlabel("Idioma")
plt.ylabel("Número de palabras")
plt.show()

avg_len_lang
```

```python
log_step("Generar nubes de palabras para reseñas positivas (>=4) y negativas (<=2)")
from wordcloud import WordCloud
import matplotlib.pyplot as plt
# --- reseñas positivas (>=4 estrellas) ---
positive_text = " ".join(df[df['review_rating'] >= 4]['review_original_text'].astype(str))

wc_pos = WordCloud(width=800, height=400, background_color="white", colormap="Blues").generate(positive_text)

plt.figure(figsize=(10,5))
plt.imshow(wc_pos, interpolation="bilinear")
plt.axis("off")
plt.title("Nube de palabras - Reseñas positivas (>=4)")
plt.show()

# --- reseñas negativas (<=2 estrellas) ---
negative_text = " ".join(df[df['review_rating'] <= 2]['review_original_text'].astype(str))

wc_neg = WordCloud(width=800, height=400, background_color="white", colormap="Reds").generate(negative_text)

plt.figure(figsize=(10,5))
plt.imshow(wc_neg, interpolation="bilinear")
plt.axis("off")
plt.title("Nube de palabras - Reseñas negativas (<=2)")
plt.show()
```

```python
log_step("Aplicar análisis de sentimiento con TextBlob y comparar con ratings en un scatterplot")
#!pip install textblob

from textblob import TextBlob

# Sentimiento de cada reseña (polarity entre -1 y 1)
df['sentiment'] = df['review_original_text'].astype(str).apply(lambda x: TextBlob(x).sentiment.polarity)

# Comparación rating vs sentimiento
plt.scatter(df['review_rating'], df['sentiment'], alpha=0.3)
plt.title("Rating vs Sentimiento de texto")
plt.xlabel("Rating (1-5 estrellas)")
plt.ylabel("Sentimiento (-1 negativo, +1 positivo)")
plt.grid(True)
plt.show()
```

```python
log_step("Calcular y graficar el sentimiento promedio por idioma")

sentiment_lang = df.groupby('review_language_requested')['sentiment'].mean().sort_values(ascending=False)

plt.figure(figsize=(8,5))
sentiment_lang.plot(kind='bar', color='seagreen', edgecolor='black')
plt.title("Sentimiento promedio por idioma")
plt.xlabel("Idioma")
plt.ylabel("Sentimiento promedio (-1 = negativo, +1 = positivo)")
plt.ylim(-1,1)
plt.show()

sentiment_lang
```

```python
log_step("Calcular y graficar el sentimiento promedio por ciudad (Top 15)")
sentiment_city = df.groupby('city')['sentiment'].mean().sort_values(ascending=False).head(15)

plt.figure(figsize=(10,5))
sentiment_city.plot(kind='bar', color='royalblue', edgecolor='black')
plt.title("Top 15 ciudades por sentimiento promedio")
plt.xlabel("Ciudad")
plt.ylabel("Sentimiento promedio (-1 = negativo, +1 = positivo)")
plt.ylim(-1,1)
plt.xticks(rotation=45, ha='right')
plt.show()

sentiment_city
```

```python
log_step("Calcular métricas clave por idioma: nº de reseñas, rating promedio, sentimiento promedio y longitud promedio")

# Agrupamos métricas clave por idioma
summary_lang = df.groupby('review_language_requested').agg(
    n_reviews=('review_text','count'),
    avg_rating=('review_rating','mean'),
    avg_sentiment=('sentiment','mean'),
    avg_length=('review_length','mean')
).round(2).sort_values('n_reviews', ascending=False)

summary_lang
```

```python
log_step("Comparar rating promedio y sentimiento promedio por idioma en gráfico de barras")

summary_lang[['avg_rating','avg_sentiment']].plot(
    kind='bar',
    figsize=(10,5),
    edgecolor='black'
)
plt.title("Rating promedio vs Sentimiento promedio por idioma")
plt.ylabel("Valor")
plt.ylim(-1,5)
plt.xticks(rotation=45, ha='right')
plt.show()
```
